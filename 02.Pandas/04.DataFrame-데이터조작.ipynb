- 결측치 처리

```python
import seaborn as sns
titanic = sns.load_dataset('titanic')
titanic.head()
```

```python
titanic.info()
```

```python
titanic.isna().sum()
```

```python
titanic.isna().sum().sum()
```

```python
titanic.tail()
```

```python
titanic.age.fillna(titanic.age.mean(), inplace=True)
titanic.tail(3)
```

```python
# embarked 는 최빈값으로 대체
titanic.embarked.value_counts()
```

```python
titanic.embarked.fillna('S', inplace=True)
titanic.tail()
```

```python
# embark_town  결측치는 행 삭제
titanic.dropna(subset=['embark_town'], inplace=True)
titanic.shape
```

```python
# deck 결측치는 너무 많아서 컬럼 삭제
titanic.drop(columns=['deck'],inplace=True)
titanic.shape
```

- apply

```python
df
```

```python
# 각 열의 최대값과 최소값의 차이
df.apply(lambda x : x.max() -x.min())
```

```python
# 각 행의 최대값과 최소값의 차이
df.apply(lambda x :  x.max() -x.min(), axis=1)
```

```python
titanic['adult/child'] = titanic.age.apply(lambda x:'adult' if x> 20 else 'child')
titanic.tail()
```

```python
# 총계 컬럼에서 천단위 구분기호를 없애고 정수로 변경
cctv['총계'] = cctv['총계'].apply(lambda x: int(x.replace(',','')))
cctv.head(3)
```

```python
cctv.head(3)
```

```python
# 결측치를 0 으로 대체, 정수로 변환
cctv.isna().sum()
```

```python
cctv['2012년'].fillna('0', inplace=True)
cctv['2013년'].fillna('0', inplace=True)
```

- 실수값을 카테고리값으로 변환

```python
age = [1,2,20,21,23,37,31,61,20,41,32,100]
bins= [1,20,30,50,70,100]
labels=['미성년자','청년','중년','장년','노년']
cats=pd.cut(age,bins,labels=labels)
cats
```
