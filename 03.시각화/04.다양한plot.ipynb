# Matplotlib - 다양한 플롯

```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
mpl.rcParams['axes.unicode_minus'] = False
mpl.rcParams['font.family'] = 'Malgun Gothic'
```

### 1. 산점도 (Scatter plot)
- x와 y의 관계를 눈으로 확인하고자 할 때

```python
X , Y = np.random.rand(100), np.random.rand(100)
plt.title('산점도')
plt.scatter(X,Y)
plt.show()
```

```python
colors = np.random.rand(100)
plt.scatter(X,Y,c=colors, alpha=0.5)
plt.title('색상을 준 산점도')
plt.show()
```

```python
area = 300 * np.random.rand(100)
plt.scatter(X,Y,s=area, alpha=0.5)
plt.title('Bubble chart')
plt.show()
```

### 2. Bar Chart

```python
s = pd.Series(np.random.rand(10), index=list('abcdefghij'))
plt.bar(s.index, s.values)
plt.title('Bar Chart')
plt.show()
```

```python
s.plot(kind='bar')
plt.show()
```

```python
import seaborn as sns
titanic = sns.load_dataset('titanic')
titanic['class'].value_counts()
```

```python
s = titanic['class'].value_counts()
plt.barh(s.index, s.values)
plt.title('타이타닉호의 선실별 승객수')
plt.show()
```

### 3. Histogram

```python
s = np.random.randn(200)
plt.hist(s)
plt.title('Historgram')
plt.show()
```

```python
s = np.random.randn(20000)
plt.hist(s, bins=20)
plt.title('Historgram')
plt.show()
```

### 4. Box plot

```python
# Setosa 품종, sepal_width
iris = sns.load_dataset('iris')
setosa = iris[iris.species == 'setosa']
```

```python
plt.boxplot(setosa.sepal_width)
plt.title('setosa품종,sepal_width')
plt.show()
```

```python
plt.boxplot([setosa.sepal_length,setosa.sepal_width], labels=iris.columns[:2])
plt.title('setosa품종,sepal_width')
plt.show()
```

- 파이(pi) 차트

```python
ratio = [ 10,15,20,25]
labels = ['감','귤','배','밤']
plt.pie(ratio, labels=labels, autopct="%.1f%%")
plt.title('Pie')
plt.show()
```

```python
colors = ['yellowgreen','gold','lightskyblue','lightcoral']
explode = (0,0,0,1,0)
plt.pie(ratio, explode=explode,colors=colors,labels=labels, autopct="%.1f%%")
plt.title('Pie')
plt.show()
```
