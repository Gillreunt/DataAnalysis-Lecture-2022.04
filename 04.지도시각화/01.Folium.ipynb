```python
# 지도위에 서울시청, 서초구청, 강남구청 표시하기

map = folium.Map(location=[37.559868, 126.966], tiles='Stamen Terrain', zoom_start=12)
folium.Marker(
    location=[37.566317,126.977829],
    popup=folium.Popup('서울특별시 중구 세종대로 110', max_width=200),
    tooltip='서울특별시청'
).add_to(map)
folium.Marker(
    location=[37.484731, 127.032078],
    popup=folium.Popup('서울특별시 서초구 남부순환로 2584(서초동) ', max_width=200),
    tooltip='서초구청',
    icon = folium.Icon(color='red', icon='info-sign')
).add_to(map)
folium.Circle(
    location=[37.517236, 127.047324],
    popup=folium.Popup('서울특별시 강남구 426 ', max_width=200),
    redius=100, color='crimson',fill=True
).add_to(map)
map
```

```python
# 지도위에 서울시청, 서초구청, 강남구청 표시하기
# tiles : stamen terrain, stamen toner , open street map(default)
map = folium.Map(location=[37.559868, 126.966], tiles='Stamen Toner', zoom_start=12)
folium.Marker(
    location=[37.566317,126.977829],
    popup=folium.Popup('서울특별시 중구 세종대로 110', max_width=200),
    tooltip='서울특별시청'
).add_to(map)
folium.Marker(
    location=[37.484731, 127.032078],
    popup=folium.Popup('서울특별시 서초구 남부순환로 2584(서초동) ', max_width=200),
    tooltip='서초구청',
    icon = folium.Icon(color='red', icon='info-sign')
).add_to(map)
folium.Circle(
    location=[37.517236, 127.047324],
    popup=folium.Popup('서울특별시 강남구 426 ', max_width=200),
    redius=100, color='crimson',fill=True
).add_to(map)
title_html = '<h3 align="center" style="font-size:20px">3개 지역 표시</h3>' 
map.get_root().html.add_child(folium.Element(title_html))
map
```
