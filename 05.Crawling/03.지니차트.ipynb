# 지니 Top 200 차트

```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
import re
```

```python
url = 'https://www.genie.co.kr/chart/top200'
req = requests.get(url)
req.text
```

```python
# Chrome User-Agent
header = {'User-Agent' : 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/100.0.4896.88 Safari/537.36'}
req = requests.get(url,headers=header)
soup = BeautifulSoup(req.text)
```

```python
trs = soup.select_one('.list-wrap').find_all('tr')
len(trs)
```

```python
trs = soup.select('.list-wrap > tbody > tr')  
len(trs)
```

```python
tr = trs[0]
```

```python
numbers = tr.select_one('.number').get_text()
numbers = numbers.split('\n')[0]
numbers
```

```python
titles = tr.select_one('.title').get_text()
titles = titles.replace('\n','').replace(' ','')
titles
```

```python
artists = tr.select_one('.artist').get_text()
artists
```

```python
albumtitles = tr.select_one('.albumtitle').get_text()
albumtitles
```

```python
gene = []
for tr in trs:
    numbers = tr.select_one('.number').get_text()
    numbers = numbers.split('\n')[0]
    titles = tr.select_one('.title').get_text()
    titles = titles.replace('\n','').replace(' ','')
    artists = tr.select_one('.artist').get_text()
    albumtitles = tr.select_one('.albumtitle').get_text()
    gene.append({'순위':numbers, '제목':titles,'가수':artists,'앨범명':albumtitles})
```

```python
df = pd.DataFrame(gene)
df
```

```python
df.to_csv('지니차트Top200.csv', index=False)
```
